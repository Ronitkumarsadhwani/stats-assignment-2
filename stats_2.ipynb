#Theory questions

Question 1: What is hypothesis testing in statistics?
 - Answer:
Hypothesis testing is a statistical method used to make decisions or inferences about a population based on a sample. It tests an assumption (the hypothesis) about a population parameter. By comparing sample data with the hypothesized value, we decide whether there is enough evidence to reject the null hypothesis in favour of the alternative hypothesis.

Question 2: What is the null hypothesis, and how does it differ from the alternative hypothesis?
 - Answer:
 - 	Null Hypothesis (H₀): A statement that assumes no effect, no difference, or status quo. Example: “The mean height of students is 160 cm.”
 - 	Alternative Hypothesis (H₁): A statement that contradicts the null hypothesis, indicating the presence of an effect or difference. Example: “The mean height of students is not 160 cm.”
The key difference is that H₀ assumes no change, while H₁ represents what we are trying to prove.

Question 3: Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.
 - vAnswer:
The significance level (α) is the probability of rejecting the null hypothesis when it is actually true. Common values are 0.05 (5%) or 0.01 (1%). It serves as a threshold:
 - If the p-value ≤ α → Reject H₀ (evidence supports H₁).
 - If the p-value > α → Fail to reject H₀ (insufficient evidence).
It balances the risk of making a Type I error.

Question 4: What are Type I and Type II errors? Give examples of each.
 - Answer:
 - 	Type I Error (False Positive): Rejecting H₀ when it is true.
Example: Concluding that a new drug is effective when it is not.
 - 	Type II Error (False Negative): Failing to reject H₀ when it is false.
Example: Concluding that a new drug is not effective when it actually is.

Question 5: What is the difference between a Z-test and a T-test? Explain when to use each.
 - Answer:
 - 		Z-test: Used when population variance is known and sample size is large (n > 30).
 - 		T-test: Used when population variance is unknown and sample size is small (n ≤ 30).
Both tests compare sample statistics to hypothesized population parameters, but the t-test accounts for greater uncertainty due to smaller samples.

```python
#Question 7: Hypothesis testing using Z-statistics in Python
import numpy as np
from scipy import stats

# Sample data
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

# Hypothesis: population mean = 50
pop_mean = 50
sample_mean = np.mean(sample_data)
sample_std = np.std(sample_data, ddof=1)
n = len(sample_data)

# Z-statistic
z = (sample_mean - pop_mean) / (sample_std / np.sqrt(n))

# p-value
p = 2 * (1 - stats.norm.cdf(abs(z)))

print("Z-statistic:", z)
print("p-value:", p)

#Interpretation:
#•	If p ≤ 0.05 → reject H₀ (sample mean differs from 50).
#•	If p > 0.05 → fail to reject H₀ (no significant difference).
```

```python
#Question 8: Normal distribution & 95% confidence interval

import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Generate normal data
data = np.random.normal(loc=50, scale=5, size=1000)

# 95% confidence interval
mean = np.mean(data)
sem = stats.sem(data)
conf_interval = stats.t.interval(0.95, len(data)-1, loc=mean, scale=sem)

# Plot
plt.hist(data, bins=30, edgecolor='black', alpha=0.7)
plt.axvline(conf_interval[0], color='red', linestyle='dashed', linewidth=2)
plt.axvline(conf_interval[1], color='red', linestyle='dashed', linewidth=2)
plt.title("Normal Distribution with 95% Confidence Interval")
plt.show()

print("95% Confidence Interval:", conf_interval)
```

```python
#Question 9: Z-scores & histogram visualization

import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Dataset
data = np.random.normal(100, 15, 1000)

# Calculate Z-scores
z_scores = stats.zscore(data)

# Plot histogram
plt.hist(z_scores, bins=30, edgecolor='black')
plt.title("Standardized Data (Z-scores)")
plt.xlabel("Z-score")
plt.ylabel("Frequency")
plt.show()

print("Z-scores calculated.")
#Explanation:
#Z-scores represent how many standard deviations a value is away from the mean:
#•	Z = 0 → at the mean
#•	Z = +1 → one standard deviation above mean
#•	Z = -1 → one standard deviation below mean
```
